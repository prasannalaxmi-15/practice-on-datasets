```python
fig, axes = plt.subplots(1, 3, figsize=(10, 5))

# Plot 1: line plot
axes[0].plot(sort_age["age"], sort_age["salary"], color="red", markersize=4)
axes[0].set_title("Age vs Salary (Line)")
axes[0].grid(True)

# Plot 2: scatter plot
axes[1].scatter(sort_age["age"], sort_age["salary"], color="green")
axes[1].set_title("Age vs Salary (Scatter)")
axes[1].grid(True)

# Plot 3: bar chart
axes[2].bar(sort_age["age"], sort_age["salary"], color="orange")
axes[2].set_title("Age vs Salary (Bar)")
axes[2].grid(True)

plt.tight_layout()
plt.show()
```
